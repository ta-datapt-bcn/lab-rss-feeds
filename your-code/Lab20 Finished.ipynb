```python
import feedparser
import pandas as pd
```

### 1. Use feedparser to parse the following RSS feed URL.

```python
url = 'http://feeds.feedburner.com/oreilly/radar/atom'
```

```python
feeds = feedparser.parse(url)
```

### 2. Obtain a list of components (keys) that are available for this feed.

```python
feeds.keys()
```

### 3. Obtain a list of components (keys) that are available for the *feed* component of this RSS feed.

```python
feeds.feed.keys()
```

### 4. Extract and print the feed title, subtitle, author, and link.

```python
print(feeds.feed.title)
print(feeds.feed.subtitle)
print(feeds.feed.link)
```

### 5. Count the number of entries that are contained in this RSS feed.

```python
dataframe = pd.DataFrame(feeds.entries)
dataframe
```

```python
entries = dataframe["title"].count()
entries
```

### 6. Obtain a list of components (keys) available for an entry.

*Hint: Remember to index first before requesting the keys*

```python
entries_key_list = list(feeds.entries[0].keys())
print(entries_key_list)
```

### 7. Extract a list of entry titles.

```python
titles = [entry["title"] for entry in feeds.entries]
print(titles)
```

### 8. Calculate the percentage of "Four short links" entry titles.

```python
four_short_links_entries = [title for title in titles if "Four short links" in title]
four_short_links_percentage = len(four_short_links_entries) / len(titles) *100
four_short_links_percentage
```

### 10. Count the number of entries per author and sort them in descending order.

```python
entries_per_author = entries_dataframe.groupby("author", as_index=False)["title"].count()
entries_per_author.columns = ["authors","entries"]
entries_per_author.sort_values("entries", ascending=False)
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
entries_dataframe["title_length"] = entries_dataframe["title"].apply(len) 
entries_dataframe[["title","author","title_length"]].sort_values("title_length", ascending=False)
```

### 12. Create a list of entry titles whose summary includes the phrase "machine learning."

```python
on_machine_learning_list = list(entries_dataframe["title"].loc[entries_dataframe["summary"].str.contains("machine learning")])
on_machine_learning_list
```
